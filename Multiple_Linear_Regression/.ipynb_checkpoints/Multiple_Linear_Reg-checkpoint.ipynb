```python
# First we deal with outliers
# Our data skewed we use IQR outlier technique
def outliers_capping(col):
    # Calculate quantile range
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1
    
    # Define Upper & Lower bounds
    Upper = Q3 + 1.5 * IQR
    Lower = Q1 - 1.5 * IQR
    
    # Capping outliers
    data = np.where(
            col > Upper,
                   Upper,
                   np.where(
                       col < Lower,
                           Lower,
                           col
            ))
    
    print('Inerquartile Range Is :', IQR)
    print('Number of Outliers are : '+ str(len(data)))
    print('Upper Limit Is :', Upper)
    print('Lower Limit Is :', Lower)
    
    return data
```

```python
# Capping outliers price column
outliers_capping(house_df['price'])
```

```python
# Capping outliers area column
outliers_capping(house_df['area'])
```

## 4. Model Building, Validation & Evaluation

```python
# Split arrays or matrices into random train and test subsets.
# Evaluate a score by cross-validation.
# Exhaustive search over specified parameter values for an estimator.
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

# Ordinary least squares Linear Regression.
# Ridge Regression.
# Lasso Regression.
# Elastic Net Regression.
# Linear model fitted by minimizing a regularized empirical loss with SGD.
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge, Lasso, ElasticNet

# Mean absolute error regression loss.
# Mean squared error regression loss.
# R2 (coefficient of determination) regression score function.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
x_scale.head()
```

```python
# Split featues into X and y
X = x_scale
y = y
```

```python
# Use train_test_split to divide the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

```python
# Call Different Regression Models
lr = LinearRegression()
rr = Ridge()
ls = Lasso()
el = ElasticNet()
sgd = SGDRegressor()
```

```python
# Store Regression models into list
models = [lr, rr, ls, el, sgd]
```

```python
# Check the testing of cross_val_score on all the models
def cross_src(model, x_trn, y_trn):
    score = np.round(cross_val_score(model, x_trn, y_trn, cv=10), 2)
    mean_src = np.round(np.mean(score), 2)
    return score, mean_src
```

```python
# Store the Testing with cross_val_score
result = []
mean_src = []

for model in models:
    current_src, current_msrc = cross_src(model, X_train, y_train)
    print('Model Name :', model)
    print('\n')
    print('Result - ',current_src)
    print('Mean_score - ',current_msrc)
    print('\n')

    result.append(current_src)
    mean_src.append(current_msrc)
```

```python
# For adjusted R2 score
# Number of rows (109)
# Total number of dependent columns (12)
X_test.shape
```

```python
# Train and test the model
def trained_model(clf,x_train,y_train,x_test,y_test):
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    mae = np.round(mean_absolute_error(y_test, y_pred), 2)
    mse = np.round(mean_squared_error(y_test, y_pred), 2)
    rmse = np.round(np.sqrt(mean_squared_error(y_test, y_pred)), 2)
    r2 = np.round(r2_score(y_test, y_pred), 2)
    adj_r2 = np.round(1-((1-r2)*(109-1) / (109-1-12)), 2)

    return mae, mse, rmse, r2, adj_r2
```

```python
# Train models
mae_src= []
mse_src= []
rmse_src = []
r2_src= []
adj_r2 = []
trn_src = []
tst_src = []

for clf in models:
    current_mae, current_mse, current_rmse, current_r2, current_adj_r2 = trained_model(clf, X_train, y_train, X_test, y_test)
    print('Name :', clf)
    print('Mean_absolute_err :\n',current_mae)
    print('Mean_squared_err :\n',current_mse)
    print('Rooted_mean_squared_err :\n',current_rmse)
    print('R2_score :\n',current_r2)
    print('Adjusted_r2_score :\n',current_adj_r2)

    mae_src.append(current_mae)
    mse_src.append(current_mse)
    rmse_src.append(current_rmse)
    r2_src.append(current_r2)
    adj_r2.append(current_adj_r2)

    # Model detail
    print('\n *** Model detail ***')
    trn = np.round(clf.score(X_train, y_train),2)
    tst = np.round(clf.score(X_test, y_test), 2)
    print(f" Training Model score :\n {trn}")
    print(f" Testing Model score :\n {tst}")
    trn_src.append(trn)
    tst_src.append(tst)
    print(f" Model coeffecient :\n {np.round(clf.coef_, 2)}")
    print(f" Model intercept :\n {np.round(clf.intercept_, 2)}")
    print('\n')
```

```python
# All models error scores are convert into dataframe
perform_df = pd.DataFrame({'Algorithm': models, 'MAE': mae_src, 'MSE': mse_src, 'Rooted MSE': rmse_src, 'R2': r2_src, 'Adjusted R2': adj_r2, 'Training Model Score': trn_src, 'Testing Model Score': tst_src})
```

```python
# Drop un-necessary cols ''
x_qwt.drop(columns=['basement_yes','hotwaterheating_yes', 'mainroad_yes'], axis=1, inplace=True)
```

```python
# Import library for VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
# VIF starts at 1 and has no upper limit
# VIF = 1, no correlation between the independent variable and the other variables
# VIF exceeding 5 or 10 indicates high multicollinearity between this independent variable and the others
Z = house_df.drop(columns=['price'], axis=1)
calc_vif(Z)
```

```python
# Feature Engineering
house_df['house_structure'] = house_df['bedrooms'] + house_df['bathrooms'] + house_df['stories'] + house_df['parking']
# Drop the cols
house_df.drop(columns=['bedrooms', 'bathrooms', 'stories', 'parking'], axis=1, inplace=True)
# Display the rows of data
house_df.head()
```

```python
# Again check multicollinearity between correlated features:
Z = house_df.drop(columns=['price'], axis=1)
calc_vif(Z)
```

```python
# Dropping one of the correlated features will help in bringing down the multicollinearity between correlated features:
Z = house_df.drop(columns=['house_structure','price'], axis=1)
calc_vif(Z)
```

```python
# Combine the correlated variables into one and drop the others. This will reduce the multicollinearity.
house_df['house'] = house_df.apply(lambda x: x['area'] - x['house_structure'], axis=1)
# Drop cols
house_df.drop(columns=['area', 'house_structure'], axis=1, inplace=True)
# Show data
house_df.head()
```

```python
# Again check multicollinearity between correlated features:
Z = house_df.drop(columns=['price'], axis=1)
calc_vif(Z)
```

```python
# Combine the correlated variables into one and drop the others. This will reduce the multicollinearity.
house_df['house_mainroad'] = house_df.apply(lambda x: x['house'] - x['mainroad_yes'], axis=1)
# Drop the col
house_df.drop(columns=['house', 'mainroad_yes'], axis=1, inplace=True)
# Top 5 rows of data
house_df.head()
```
